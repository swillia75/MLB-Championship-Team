```python
#import numpy to use seed 
import numpy as np
import pandas as pd
from numpy.random import seed
np.random.seed(1)
```

Read CSV create, and clean dataframe

```python
#Read in CSV and create dataframe
mlb_teams= pd.read_csv("data/Teams.csv")
mlb_teams= mlb_teams.dropna()

mlb_teams
```

```python
#Get list of dataframe columns

mlb_teams.columns
```

```python
#Select columns for dataframe

mlb_df = mlb_teams[['W', 'L', 'WPCT', 'WSWin', 'R', 'AB', 'H',
       '2B', '3B', 'HR', 'BB', 'SO', 'SB', 'ERA', 'SHO', 'SV', 'name']]
mlb_df.head(20)
```

```python
#Use getdummies to convert strings

data_LG = mlb_df.copy()
data_bin = pd.get_dummies(data_LG, columns=["WSWin"])
data_bin = data_bin.drop(columns=['WSWin_N'])

data_bin.head()
```

```python
data = data_bin.drop("name", axis = 1)
outcome = data_bin["name"]
print(data.shape, outcome.shape)
```

Preprocess data

```python
#Import SciKit libraries

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.utils import to_categorical
```

```python
#Split data and teams into training and test data

X_train, X_test, y_train, y_test = train_test_split(data, outcome, random_state=42)
y_train
```

```python
#Scale data using StandardScaler

X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
X_train_scaled
```

Create and train model

```python
#Import and use SVC MODEL
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)
```

```python
#Print training and test scores from model using datasets

print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [100, 500, 1000],
              'gamma': [0.0001, 0.005, 0.001]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
# Train the model with GridSearch

grid.fit(X_train_scaled, y_train)
```

Evaluate Model

```python
#Print best parameters and score from Gridsearch

print(grid.best_params_)
print(grid.best_score_)
```

```python
#Print accuracy of model using test dataset

print('Test Acc: %.3f' % grid.score(X_test_scaled, y_test))
```

Predict World Series Champs

```python
#Preprocess and predict 2015 MLB WS Champ
mlb_teams2015= pd.read_csv("data/2015.csv")

mlb2015_df = mlb_teams2015[['W', 'L', 'WPCT', 'WSWin', 'R', 'AB', 'H',
       '2B', '3B', 'HR', 'BB', 'SO', 'SB', 'ERA', 'SHO', 'SV', 'name']]

data2015_LG = mlb2015_df.copy()
data2015_bin = pd.get_dummies(data2015_LG, columns=["WSWin"])
data2015_bin = data2015_bin.drop(columns=['WSWin_N'])

feature2015 = data2015_bin.drop("name", axis=1)
team2015 = data2015_bin["name"]

X_feature2015_scaler = StandardScaler().fit(feature2015)
X_feature2015_scaled = X_feature2015_scaler.transform(feature2015)

predictions2015 = grid.predict(X_feature2015_scaled)
print('Test Acc: %.3f' % grid.score(X_feature2015_scaled, team2015))
predictions2015 = model.predict(X_feature2015_scaled)
print(predictions2015)
```
