```python
#import numpy to use seed 
import numpy as np
import pandas as pd
from numpy.random import seed
np.random.seed(1)
```

```python
#import other dependencies

import os
import tensorflow 
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense
```

Read CSV create, and clean dataframe

```python
#Read in CSV and create dataframe
mlb_teams= pd.read_csv("Teams.csv")

mlb_teams
```

```python
#Get list of dataframe columns

mlb_teams.columns
```

```python
#Select columns for dataframe

mlb_df = mlb_teams[['yearID', 'lgID', 'W', 'L', 'WSWin', 'R', 'AB', 'H', '2B',
       '3B', 'HR', 'BB', 'SO', 'SB', 'ERA', 'CG', 'SHO', 'SV', 'name']]


mlb_df.tail(37)
```

```python
#Use getdummies to convert strings

data_LG = mlb_df.copy()
data_bin = pd.get_dummies(data_LG, columns=["WSWin", 'lgID'])
data_bin = data_bin.drop(columns=['WSWin_N', 'lgID_AL'])

data_bin.head()
```

```python
#Set X(data) and y(team)variables

data = data_bin.drop("name", axis = 1)
outcome = data_bin["name"]
print(data.shape, outcome.shape)
```

Preprocess data

```python
#Import SciKit libraries

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.utils import to_categorical
```

```python
#Split data and teams into training and test data

X_train, X_test, y_train, y_test = train_test_split(data, outcome, test_size = 0.2, random_state=42)
y_train
```

```python
#Scale data using MinMaxScaler

X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
X_train_scaled
```

```python
##Import RandomForestClassifier and train
#from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000, random_state=1)
#classifier = MultiOutputClassifier(rf, n_jobs=-1)
rf.fit(X_train_scaled, y_train)
```

```python
#Get Score for RandomForestClassifier
rf.score(X_test_scaled, y_test)
```

```python
#Print data scores for training aand test datasets
print(f"Training Data Score: {rf.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {rf.score(X_test_scaled, y_test)}")
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {"bootstrap": [True],
              "max_depth": [20, 40, 100, 125],
#              "max_features": [2,3],
 #            "min_samples_leaf": [3, 4, 5],
              "n_estimators": [100, 500, 1000]
              }
grid = GridSearchCV(rf, param_grid, verbose=3)
grid
```

Train Model

```python
# Train the model with GridSearch

grid.fit(X_train_scaled, y_train)
```

```python
#Print best parameter and score 

print(grid.best_params_)
print(grid.best_score_)
```
