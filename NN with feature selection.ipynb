```python
#import numpy to use seed 
import numpy as np
import pandas as pd
from numpy.random import seed
np.random.seed(42)
```

```python
#import other dependencies

import os
import tensorflow 
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense
```

Read CSV create, and clean dataframe

```python
#Read in CSV and create dataframe
mlb_teams= pd.read_csv("Teams.csv")
mlb_teams = mlb_teams.dropna()
mlb_teams
```

```python
#Get list of dataframe columns

mlb_teams.columns
```

```python
#Select columns for dataframe


mlb_df = mlb_teams[['W', 'L', 'WPCT', 'WSWin', 'R', 'H', '2B',
       '3B', 'HR', 'BB', 'SB', 'ERA', 'SHO', 'SV', 'name']]
mlb_df.head(20)
```

```python
#Use getdummies to convert strings

data_LG = mlb_df.copy()
data_bin = pd.get_dummies(data_LG, columns=["WSWin"])
data_bin = data_bin.drop(columns=['WSWin_N'])

data_bin.head()
```

```python
#Set X(data) and y(team)variables

data = data_bin.drop("name", axis=1)
outcome = data_bin["name"]
print(data.shape, outcome.shape)
```

```python
outcome.unique()
```

Preprocess data

```python
#Import SciKit libraries

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.utils import to_categorical
```

```python
#Split data and teams into training and test data

X_train, X_test, y_train, y_test = train_test_split(data, outcome, train_size=0.6, test_size=0.4, random_state=1)
X_train
```

```python
#Scale data using StandardScaler

X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
X_train_scaled
```

```python
#Label-encode team

label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
encoded_y_train
```

```python
#One hot_encode the team data

y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
for x in y_test_categorical:
    print(x)
```

Create Neural Network Model

```python
# Create model and add layers
model = Sequential()
model.add(Dense(100, input_dim=14, activation='relu'))

model.add(Dense(36, activation='softmax'))
```

```python
#View model summary

model.summary()
```

Train Model

```python
#Compile model
model.compile(optimizer="adam",
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
print(len(y_train_categorical))
```

```python
#Fit data to train model

model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

Evaluate Model

```python
#Evaluate model

model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(f"Deep Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
